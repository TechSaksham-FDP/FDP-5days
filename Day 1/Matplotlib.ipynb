#### 1. Line Plot

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

#### 2. Scatter Plot

Pandas is a software library written for the Python programming language for data manipulation and analysis.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.randint(100, size = (100))
y = np.random.randint(100, size = (100))
colors = np.random.randint(100, size = (100))
sizes = 10 * np.random.randint(100, size = (100))
plt.figure(figsize = (5,5))
plt.scatter(x,y,c=colors,s=sizes, alpha = 0.8, cmap = 'prism')
```

##### 3. Bar Chart

- bar chart is basically used for see the comparision between various parameters

```python
company=["TCS","Infy","LTI","PSL"]
revenue=[400,370,90,160]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(company,revenue,label="Revenue")
plt.xlabel("Company")
plt.ylabel("Revenue in Bill. Dollars")
plt.title("Graph for company vs Revenue")
plt.legend()
```

##### 4. Pie Charts

- it used to plot the graph with percentages for each criteria. for ex company expenses

```python
company_expenses=[100,60,40,20,80]
expense_data=["emp_sal","IF Devlopment","Food Expenses","Travelling Allow","prizes"]
```

##### 5. Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.normal(170,10,250)
plt.hist(x, bins = 20, color = 'red')
plt.show()
```

#### Legends and Annotations

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.0,5.0,0.001)
s = np.cos(3*np.pi*t)
line = plt.plot(t,s,lw = 2)
plt.annotate('Local max', xy = (3.3,1),
            xytext = (3,1.8),
            arrowprops = dict(facecolor = 'green', shrink = 0.05))
plt.ylim(-2,2)
plt.show()
```

### 3D Plot

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.axes(projection='3d')
z = np.linspace(0,20,2000)
x = np.sin(z)
y = np.cos(z)
fig.plot3D(x,y,z,'blue')
fig.set_title('3d plot')
fig.plot3D(x,y,z,'blue',linewidth=5,linestyle = '--')
```
