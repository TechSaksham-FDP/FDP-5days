# Pandas Series

Pandas Series data structure is a one-dimensional labelled array, holding data of any type. It is the primary building block for a DataFrame, making up its rows and columns.

```python
import pandas as pd
a = [10,"Hello", 27.5, True]
myseries = pd.Series(a)
print(myseries)
```

# Labels

If nothing is specified, the values are labeled with their index number. First value has index 0, seconf value has index 1 etc.

```python
print(myseries[1])
```

```python
#with the index argument, you can name your own labels

import pandas as pd
a = [10,"Hello", 27.5, True]
myseries = pd.Series(a, index = ['A','B','C','D'])
print(myseries)
```

```python
#After creating labels, you can access the item by referring to the label

print(myseries['A'])
```

# DataFrames

The Pandas DataFrame is a two-dimensional data structure composed of columns and rows. You can think of the DataFrame as similar to a CSV or relational database table.  

Series is like a column, a DataFrame is the whole table

```python
# creating a dataframe from two Series

import pandas as pd

result = {
        'Subjects':['Maths','Physics','Chemistry','Biology','English','Hindi'],
        'Marks':[86, 73, 91, 82, 75, 78]
        }

df = pd.DataFrame(result)
print(df)
```

```python
# creating dataframe with the index argument
import pandas as pd
data = {
            "Name": [
                "Mr. Edward Bell",
                "Miss. Priyanka Paul",
               "Miss. Elizabeth",
           ],
            "Age": [22, 30, 58],
            "Sex": ["male", "female", "female"],
        }
df = pd.DataFrame(data, index = ['Passenger 1', 'Passenger 2', 'Passenger 3'])
print(df)
```

```python
# Accessing Column
df["Age"]
```

```python
# use the loc attribute to return the specified row(s)
print(df.loc['Passenger 2'])
```

```python
print(df.loc['Passenger 2']['Name'])
```

```python
# Finding Max value from Column
df["Age"].max()
```

```python
# Statitical Analysis of Column
df.describe()
```

# Plot a Line Plot using Pandas

```python
# Prepare the data and create the DataFrame

import pandas as pd
   
df = pd.DataFrame({
    'Subject_1': [70.5, 80.7, 50.4, 70.5, 80.9],
    'Subject_2': [40.24, 50.9, 70.6, 80.1, 50.9],
    'Subject_3': [30, 50.5, 70.8, 90.88, 30],
    'Exam' : ['Exam_1','Exam_2','Exam_3','Exam_4','Exam_5']
                 })
print(df)
```

```python
# Line Plot using 'plot' function on dataframe
df.plot()
```

```python
df.plot(x = 'Exam', y = 'Subject_2')
```

# Area Plot

Extended version of the line plot. In the area plot, instead of leaving the plot to connected data points, the area under the line is filled with colors. This is helpful in cases when you want to show the proportion of values captured by a particular value. 

```python
df = pd.DataFrame(np.random.rand(30, 4),
       columns =['A', 'B', 'C', 'D'])
df.plot.area();
```

# Scatter Plot

```python
df = pd.DataFrame(np.random.rand(200, 2),
           columns =['Num 1', 'Num 2'])
df.plot.scatter(x ='Num 1', y ='Num 2', color = 'r');
```

# Bar Plot

```python
# Prepare the data and create the DataFrame

import pandas as pd
   
data = {'Country': ['USA','Canada','Germany','UK','France'],
        'GDP_Per_Capita': [45000,42000,52000,49000,47000]
       }
  
df = pd.DataFrame(data,columns=['Country','GDP_Per_Capita'])
df.plot.bar(x = 'Country', y = 'GDP_Per_Capita', color = 'orange')
```

```python
df.plot.barh(x = 'Country', y = 'GDP_Per_Capita', color = 'orange')
```

# Histogram

```python
df = pd.DataFrame(
               np.random.randint(1,7,6000),
               columns = ['one'])
df['two'] = df['one'] + np.random.randint(1,7,6000)
ax = df.plot.hist(bins = 12, alpha = 0.7) #alpha parameter defines the transparency color
```

# Pie Chart

```python
# Prepare the data and create the DataFrame

import pandas as pd

data = {'Tasks': [300,500,700]}
df = pd.DataFrame(data,columns=['Tasks'],index = ['Tasks Pending','Tasks Ongoing','Tasks Completed'])

print (df)
```

```python
# Plot the DataFrame using Pandas

import pandas as pd
import matplotlib.pyplot as plt

data = {'Tasks': [300,500,700]}
df = pd.DataFrame(data,columns=['Tasks'],index = ['Tasks Pending','Tasks Ongoing','Tasks Completed'])

df.plot.pie(y='Tasks',figsize=(5, 5),autopct='%1.1f%%', startangle=90)
plt.show()
```
